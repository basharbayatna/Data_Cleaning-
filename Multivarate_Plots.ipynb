```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np



url='/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data-tittan-cleaned.csv'
df = pd.read_csv(url)
```

#Q1) Plot The counts for the Number of Passengers  by sex if they survived or not

```python
#sns.barplot(data=df, y='Survived', x='Sex')
ax = sns.countplot(data=df, x='Survived', hue='Sex',palette='Set1')
ax.set(title='Passengers Status By Sex')
```

- The females with almost 250 survivors

#Q2) Barplot the AVG number of 'parents or Children' for those that survived vs those that died

```python
ax = sns.barplot(data=df, y='Parents', x= 'Survived',palette='Set1')
ax.set(title='AVG # of Parents')
```

-We can notice that the survived Group has a  slightly higher AVG

#Q3) Plot a Correlation Heatmap for all the numaric columns

```python
corr = df.corr(numeric_only=True)
corr.round(2)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
sns.heatmap(corr,cmap='Blues', annot=True)
ax.set(title='Correlation heatmap')
```

-The Fare has The Highest Positive Correlation to the Survived Column
